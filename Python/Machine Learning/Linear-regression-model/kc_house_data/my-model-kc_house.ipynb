## Predict House Sales in King County, USA prices
#### Goal: Predict house price using regression. The label is the "price" column.

Data source: https://www.kaggle.com/datasets/harlfoxem/housesalesprediction

```python
# Import of libraries and visual analysis of data

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


pd.set_option('display.max_columns', 21)
dataset = pd.read_csv(r"./kc_house_data/kc_house_data.csv")
display(dataset.head()); print(dataset.info())
```

```python
# Defining the label (dataset_y) and features (dataset_x)
# After analyzing the dataset, the treatments'll already be
# carried out (in this case it'll be enough to just remove some columns)

dataset_y = dataset["price"]; display(dataset_y.head()) # var targer

colum_drop = ["id", "price", "date", "condition", "sqft_basement"]
dataset_x = dataset.drop(columns=colum_drop); display(dataset_x.head()) # var predictors
```

```python
# Split test and training data
# test_size=0.25, it means 25% of the data'll be used for testin' and 75% will be used for trainin'
# random_state=10, it means the data scrambling'll be the same for all executions
x_treino, x_teste, y_treino, y_teste =  train_test_split(dataset_x, dataset_y, test_size=0.25, random_state=10)
```

```python
# Creation and execution of the machine learning linear regression model with the inserted training data
model = LinearRegression(); model.fit(x_treino, y_treino)
```

```python
# Results for the train data and especially for the test data (using the coefficient of determination R²)
result_x = model.score(x_treino, y_treino); print(f">> r² (training data): {result_x:.2%}")
result_y = model.score(x_teste, y_teste); print(f">> r² (test data): {result_y:.2%}")
```

```python
# Making a prediction with the created model

to_predict = x_teste.loc[957:957]; print(">> chosen line 957 for random prediction:"); display(to_predict)

print(">> Price forecast by model: ${:.1f}".format(model.predict(to_predict)[0]))
print(">> Price set by default: ${}\n".format(dataset["price"][957]))
print(">> Accuracy of this prediction: {:.2%}\n>> Error distance: ${:.2f}".format(
    (dataset["price"][957] / model.predict(to_predict)[0]),
    (model.predict(to_predict)[0] - dataset["price"][957])),
)

print("\n>> line 957 original:"); display(dataset.loc[ dataset["id"] == 252000400])
```
